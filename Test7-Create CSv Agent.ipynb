```python
from langchain.agents.agent_types import AgentType
from langchain_experimental.agents.agent_toolkits import create_csv_agent
from langchain_openai import ChatOpenAI, OpenAI

llm= ChatOpenAI(model='gpt-4o',temperature=0)
file_path=["./data/Obstructive Lung Diseases - Copy.csv"]
agent = create_csv_agent(
    llm=llm,
    path=file_path,
    verbose=True,
    agent_type=AgentType.OPENAI_FUNCTIONS,
    allow_dangerous_code=True,
    handle_parsing_errors=True
)

query="""Can you rank the top 5 countries Based on the scenario where patient population is weighted at 50% 
and competition weighted at 50%?"""
agent.invoke(query)
```

```python
from langchain_experimental.agents.agent_toolkits import create_pandas_dataframe_agent
from langchain_openai import ChatOpenAI
import pandas as pd

# Initialize the language model
llm = ChatOpenAI(model='gpt-4-0613', temperature=0)

# Read the CSV file
df = pd.read_csv("./data/Obstructive Lung Diseases (3).csv")

# Create the pandas dataframe agent
agent = create_pandas_dataframe_agent(llm, df, verbose=True,allow_dangerous_code=True,)

# Function to handle user queries
def process_query(query):
    try:
        result = agent.run(query)
        return result
    except Exception as e:
        return f"An error occurred: {str(e)}"

# Example usage
query = """Can you rank the top 5 countries Based on the scenario where patient population is weighted at 50% 
and competition weighted at 50%?"""
query="""Can you weigh Patient Population at 25%, Competition at 25%, 
and Country Operations at 50% for Obstructive Lung Diseases to give me the prioritized country list? Please in proper format"""
query="""
Can you weigh Patient Population at 90%, Competition at 5%, and Country Operations at 5% for Obstructive Lung Diseases to give me the prioritized country list? Give the output in proper format with numbers."""
query="Can you prioritize the countries with 50% weight for Patient Incidence, 25% weight for Roche Recruitment rate, 25% weight of sites with no competitor trials? Provide the answer in a table"

response = process_query(query)
print(response)

# You can now use the process_query function with any query related to the CSV
```
